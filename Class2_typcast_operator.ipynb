## Typecasting

Typecasting in Python, also known as type conversion, refers to converting one data type into another. Python provides several built-in functions to facilitate typecasting.

### Types of Typecasting

1.   Implicit Typecasting
2.   Explicit Typecasting

image.png

#### Implicit Typecasting



Implicit typecasting occurs automatically when Python converts one data type to another without the need for explicit intervention by the user. This usually happens when performing operations between different data types.

```python
x = 5    # Integer
y = 3.0  # Float

result = x + y  # x is implicitly converted to float
print(result)   # Output: 8.0
print(type(result))  # Output: <class 'float'>
```

#### Explicit Typecasting

Explicit typecasting is performed when the programmer explicitly converts one data type to another using built-in functions. This is also known as type conversion.

**Common Typecasting Functions**

* int(): Converts a value to an integer.
* float(): Converts a value to a float.
* str(): Converts a value to a string.
* list(): Converts a value to a list.
* tuple(): Converts a value to a tuple.
* set(): Converts a value to a set.
* dict(): Converts a value to a dictionary (from a list of key-value pairs).
* bool(): Converts a value to a boolean.

### Difference between ~ and ! in

**Bitwise NOT Operator (~)**

* Usage: The ~ operator is a bitwise NOT operator. It inverts all the bits of an integer.

* Operation: It flips the bits of the number, which means it changes 0 bits to 1 and 1 bits to 0.

```python
a = 5  # binary representation: 0000 0101
result = ~a  # binary representation after NOT: 1111 1010
print(result)  # result is -6 (in decimal)
```

**Explanation:** The ~ operator flips the bits of 5, and because Python uses signed integers with two's complement representation, the result is -6

**Logical NOT Operator (!)**

* Usage: In Python, the logical NOT operator is not !; it is represented by not. The not operator is used to invert the boolean value of an expression.

* Operation: It changes True to False and False to True.


```python
a = True
result = not a  # result is False
print(result)
```

**Explanation:** The not operator inverts the boolean value of True to False.

**Key Differences:**
- Operation Type:

 - ~: Bitwise operation.
 - not: Logical operation.
- Operand Type:

  - ~: Operates on integers.
  - not: Operates on boolean values or expressions.

- Output:

  - ~: Returns an integer.
  - not: Returns a boolean.







## **Introduction to Binary Number System**

### **What is the Binary Number System?**

- The binary number system is a base-2 numeral system that uses only two symbols: 0 and 1.
- It is the foundation of digital computing.

- **Place Values:** Each digit represents a power of 2.

- For example, the binary number 1011 represents:

- $1 * 2^3 + 0 * 2^2 + 1 * 2^1 + 1 * 2^0$
- Which is equal to $8 + 0 + 2 + 1 = 11$ in decimal.

###**Binary to Decimal Conversion**

- Convert a binary number to decimal by summing the products of each bit and its corresponding power of 2.

Example:
- Binary: 1101
- Decimal Calculation : $ 1*2^3+1*2^2+0*2^1+1*2^0 = 8+4+0+1 = 13 $

```python
def binary_to_decimal(binary_str):
    return int(binary_str, 2)

# Example usage
binary_str = "1100001000001"
decimal_value = binary_to_decimal(binary_str)
print(f"Decimal value: {decimal_value}")
```

```python
def binary_to_decimal_logical(binary_str):
    """Converts a binary string to its decimal equivalent using logical programming.

    Args:
        binary_str: The binary string to convert.

    Returns:
        The decimal equivalent of the binary string.
    """

    decimal_value = 0
    power = 0

    for digit in reversed(binary_str):
        if digit == '1':
            decimal_value += 2 ** power
        power += 1


    return decimal_value

# Example usage
binary_str = "1100001000001"
decimal_value = binary_to_decimal_logical(binary_str)
print(f"Decimal value: {decimal_value}")  # To see the output, run the code.
```

###**Decimal to Binary Conversion**

- Convert a decimal number to binary by repeatedly dividing by 2 and recording the remainders.

**Example:**
- Decimal: 13
- Binary Calculation: Divide 13 by 2 repeatedly, recording remainders. The binary representation is 1101.

```python
def decimal_to_binary(decimal_num):
    return bin(decimal_num).replace("0b", "")

# Example usage
decimal_num = 13
binary_str = decimal_to_binary(decimal_num)
print(f"Binary value: {binary_str}")
```

###**Binary Addition**

- Add binary numbers by aligning them and performing bit-wise addition, carrying over if needed.

**Example:**
- Binary Addition: 1010 + 1101

image.png

```python
def binary_addition(bin1, bin2):
    return bin(int(bin1, 2) + int(bin2, 2)).replace("0b", "")

# Example usage
result = binary_addition("1010", "1101")
print(f"Binary addition result: {result}")
```

###**Binary Subtraction**

- Subtract binary numbers by aligning them and performing bit-wise subtraction, borrowing if needed.

**Example:**
- Binary Subtraction: 1101 - 1010


image.png

```python
def binary_subtraction(bin1, bin2):
    return bin(int(bin1, 2) - int(bin2, 2)).replace("0b", "")

# Example usage
result = binary_subtraction("1101", "1010")
print(f"Binary subtraction result: {result}")
```

###**Binary Multiplication**

- Multiply binary numbers using bit-wise multiplication and summing the partial results.

**Example:**
- Binary Multiplication: 101 * 11

image.png

```python
def binary_multiplication(bin1, bin2):
    return bin(int(bin1, 2) * int(bin2, 2)).replace("0b", "")

# Example usage
result = binary_multiplication("101", "11")
print(f"Binary multiplication result: {result}")
```

###**Binary Division**

- Divide binary numbers using bit-wise division, similar to long division in decimal.

**Example:**
- Binary Division: 1010  10

```python
def binary_division(bin1, bin2):
    return bin(int(bin1, 2) // int(bin2, 2)).replace("0b", "")

# Example usage
result = binary_division("1010", "10")
print(f"Binary division result: {result}")
```
