## 2.Iterative Statements

Iterative statements, also known as loops, allow you to execute a block of code repeatedly based on a condition. Python provides several types of iterative statements to handle different looping needs: for loops, while loops, and control statements like continue, break, and pass.

###Types of Iterative Statements

* for Loop
* while Loop


#### 2.1for loop

The for loop in Python is used to iterate over a sequence (such as a list, tuple, string, or range) and execute a block of code for each item in the sequence. It is one of the most commonly used iterative statements in Python.

Syntax


```
for variable in sequence:
    # code to execute for each item in the sequence


```



image.png

**Important Points**

* Sequence: The for loop can iterate over any sequence such as lists, tuples, strings, and ranges.
* Indentation: The code block inside the for loop must be indented.
* range() Function: The range() function is commonly used with for loops to generate a sequence of numbers.
* enumerate() Function: The enumerate() function can be used to get both the index and the item during iteration.

**Example 1:**

**2.1.1Using range() Function**
   - The range() function generates a sequence of numbers, which is particularly useful in for loops when you need to iterate a specific number of times.

#### 2.2While loop

The while loop in Python repeatedly executes a block of code as long as a specified condition is true. It's useful for scenarios where the number of iterations is not known beforehand and depends on dynamic conditions during runtime.
```
Syntax

while condition:
    # code to execute as long as the condition is true
```





image.png

**Important Points**
* Condition: The condition in the while loop is evaluated before each iteration. If the condition is true, the code block is executed. If the condition is false, the loop terminates.
* Infinite Loop: Be cautious with while loops as they can easily lead to infinite loops if the condition never becomes false. Using break statements can help prevent this.
* Indentation: The code block inside the while loop must be indented properly.

Example 1: Print 0 to n-1 using while loop

````python
count = 0
while count < 5:
    print(count)  # Output: 0 1 2 3 4
    count += 1
````

Example 2: Infinite while loop

````python
while True:
    print("This is an infinite loop")
    break  # This line breaks the loop to prevent it from running indefinitely
````

**Example 3:**
- **User Input Validation**

````python
password = ''
while password != 'secret':
    password = input("Enter the password: ")
print("Access granted!")
````

## 3.Transfer statements

Transfer statements in Python are used to alter the flow of execution within loops or switch between different parts of the code. They control when to exit the loop, skip an iteration, or do nothing (serve as a placeholder). The main transfer statements are break, continue, and pass.

**Types of Transfer Statements**
1. break Statement
2. continue Statement
3. pass Statement

image.png

### Continue

The continue statement is used within loops to skip the rest of the code inside the current iteration and proceed to the next iteration of the loop. It is commonly used to avoid executing **certain code** under specific conditions.

**Syntax**


```
continue
```




image.png

#### **Using continue in for loop**

**Example: Skipping Even Numbers**

````python
for i in range(10):
    if i % 2 == 0:
        ##print(f"continue skipped this {i}")
        continue
    print(i)
# Output: 1 3 5 7 9
````

````python
for i in range (10,90):
  if i%5!=0:
    continue
  print(i)
````

````python
for i in range (10,90):
  if i==13:
    break
  print(i)
````

**Exmaple: Skipping Specific Values**

````python
for letter in 'Python':
    if letter == 't':
        continue
    print(letter)
# Output: P y h o n
````

#### **Using continue in a while Loop**

````python
count = 0
while count < 10:
    count += 1
    if count % 2 == 0:
        continue
    print(count)
# Output: 1 3 5 7 9
````

**Important Points**
* Scope: The continue statement affects only the innermost loop in which it is placed. In nested loops, it will skip the rest of the code in the current iteration of the inner loop.
* Effect: When continue is executed, the rest of the code inside the loop block is skipped for the current iteration, and the loop proceeds with the next iteration.
* Usage: The continue statement is useful for avoiding execution of code based on certain conditions, reducing the need for nested if statements.

### Break

The break statement is used to exit a loop prematurely, regardless of the loop's condition. It immediately terminates the loop and transfers control to the statement following the loop.

**Syntax**



```
break
```



image.png

#### **Exiting a for Loop**

````python
for i in range(10):
    if i == 5:
       print(f"break executed at {i}")
       break
    print(i)
print("for loop executed")
# Output: 0 1 2 3 4
````

````python
for i in range (10):
  if i ==5:
    continue
  if i ==9:
    break
  print(i)
````

#### **Exiting a while Loop**

````python
count = 0
while count < 10:
    print(count)
    if count == 7:
        break
    count += 1
# Output: 0 1 2 3 4 5 6 7
````

````python
s = 'python'
count =0

while count<10:
  for i in s:
    if i =='t':
      break
    print(i)
  count+=1
  print(count)


````

**Important Points**

* Scope: The break statement affects only the innermost loop where it is placed. It does not exit outer loops or functions.

* Termination: When break is executed, the loop terminates immediately, and the program control moves to the next statement following the loop.

* Usage: The break statement is useful for exiting loops based on specific conditions, avoiding unnecessary iterations and improving code efficiency.

### Pass

The pass statement in Python is a null operation; it is used as a placeholder where syntactically some code is required but where no action is needed or intended. It is often used in code where a statement is syntactically required but where no action needs to be performed.

Syntax


```
pass
```




#### **Placeholder in Functions**

````python
def placeholder_function():
    pass  # Function is not yet implemented

# Calling the function
placeholder_function()
````

#### **Placeholder in Loops**

````python
for i in range(5):
    pass  # No action is performed in the loop

print("Loop completed.")
````

#### **Placeholder in Classes**

````python
class MyClass:
    pass  # Class is not yet implemented

# Creating an instance of the class
obj = MyClass()
````

**Important Points**

* Placeholder: pass is often used as a placeholder during development to indicate where code will be added later.
* Syntax Requirement: It is used where a statement is syntactically required but no action is needed.
* No Effect: The pass statement does nothing and is used only to maintain code structure.

## 4.**Loop Control with else**

- In Python, loops can have an else clause, which executes after the loop completes normally.
- **However, it does not execute if the loop is terminated by a break statement.**

**Example:**
- **Loop with else**

````python
for num in range(5):
    print(num)
else:
    print("Loop completed!")
````

**Example:**
- **else with break**

````python
for num in range(5):
    if num == 3:
        break
    print(num)
else:
    print("Loop completed!")
````

- In this example, the else block does not execute because the loop is terminated by the break statement.

### Problem 3: Sum of Prime Numbers in a Range

**Problem Statement:**
- Write a Python program to find the sum of all prime numbers between two given numbers a and b (inclusive).
- Use a for loop to iterate through the range and check for prime numbers.
- Use the break statement to exit the inner loop when a non-prime condition is met.

````python
import math

def is_prime(n):

  if n <= 1:
    return False
  for i in range(2, int(math.sqrt(n)) + 1):
    if n % i == 0:
      return False
  return True

# Example usage
number = 25
if is_prime(number):
  print(f"{number} is a prime number")
else:
  print(f"{number} is not a prime number")
````

````python
# Input range from the user
a = int(input("Enter the starting number of the range: "))
b = int(input("Enter the ending number of the range: "))

import math

# Calculate the sum of prime numbers in the given range
prime_sum = 0

for num in range(a, b + 1):
    is_prime = True
    if num <= 1:
        continue
    for i in range(2,int(math.sqrt(num)) + 1):
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
      prime_sum +=num
      print(num)

print(f"The sum of prime numbers between {a} and {b} is: {prime_sum}")
````

1### **Problem 4: Counting Vowels in a String**

**Problem Statement:**
- Write a Python program to count the number of vowels in a given string using a for loop and an if statement.

````python
# Input string from the user
input_string = input("Enter a string: ")

# Initialize vowel count to 0
vowel_count = 0
vowels = "aeiouAEIOU"

# Count the number of vowels in the string
for c in input_string:
    if c in vowels:
        vowel_count += 1

print(f'The number of vowels in the string is: {vowel_count}')
````

````python
##count of each vowel

str ='abcdeeooujhjmnlo'
counta,counte =0,0

for i in str:
  if i =='a':
    counta +=1
  if i =='e':
    counte +=1
print(f"number of a = {counta} ,number of e = {counte}")
````

````python
##dictionary

str_input = 'abcdeeooujhjmnlo'

# Initialize a dictionary to count vowels
vowel_counts = {vowel: 0 for vowel in 'aeiou'}

# Iterate through the string
for c in str_input:
    if c in vowel_counts:  # Check if the character is a vowel
        vowel_counts[c] += 1

# Print the counts for each vowel
for vowel, count in vowel_counts.items():
    print(f"Number of {vowel} = {count}")
````
